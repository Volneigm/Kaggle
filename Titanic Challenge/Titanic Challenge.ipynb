```python
# Import libraries 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# Read data
train =  pd.read_csv('train.csv')
test =  pd.read_csv('test.csv')
```

```python
# View dataframes
train.head()
```

```python
# Drop columns 'name', 'ticket', 'cabin'
train.drop(['Name', 'Cabin', 'Ticket'], axis = 1, inplace = True)
test.drop(['Name', 'Cabin', 'Ticket'], axis = 1, inplace = True)

# View the new  dtaset
train.describe()
```

```python
train.head()
```

```python
test.head()
```

```python
# One Hot Encoding not number values (categorical values)
new_data_train = pd.get_dummies(train)
new_data_test = pd.get_dummies(test)
```

```python
new_data_train.head()
```

```python
# Visualise NaN  train values
new_data_train.isnull().sum().sort_values(ascending=False)
```

```python
# Fill NaN values in both datasets
new_data_train['Age'].fillna(new_data_train['Age'].mean(), inplace=True)
new_data_test['Age'].fillna(new_data_test['Age'].mean(), inplace=True)
```

```python
# Visualise NaN test values
new_data_test.isnull().sum().sort_values(ascending=False)
```

```python
# Fill NaN values in test Dataset
new_data_test['Fare'].fillna(new_data_test['Fare'].mean(), inplace=True)
new_data_test.head()
```

```python
# Creating features and target (Create a model)
X = new_data_train.drop('Survived', axis =1)
y = new_data_train['Survived']
```

```python
# Create the train model
tree = DecisionTreeClassifier(max_depth=3, random_state=0)
tree.fit(X, y)
```

```python
pred = tree.predict(new_data_test)
print(pred)
tree.score(X,y)
```

```python
submission = pd.DataFrame()
submission['PassengerId'] = new_data_test['PassengerId']
submission['Survived'] = tree.predict(new_data_test)
submission.to_csv('submission.csv', index=False)
```
